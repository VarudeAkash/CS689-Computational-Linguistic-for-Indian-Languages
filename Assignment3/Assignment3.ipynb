```python
import torch
from indicnlp.tokenize import indic_tokenize
from nltk.translate.bleu_score import corpus_bleu
from rouge_score import rouge_scorer
from IndicTransTokenizer import IndicProcessor, IndicTransTokenizer
from transformers import AutoModelForSeq2SeqLM
from indicnlp.transliterate.unicode_transliterate import ItransTransliterator
import os
from tqdm import tqdm
```

##### English to marathi translation

```python
def indictranslate(src, tgt, sents):
    tokenizer = IndicTransTokenizer(direction="en-indic")
    ip = IndicProcessor(inference=True)
    model = AutoModelForSeq2SeqLM.from_pretrained("ai4bharat/indictrans2-en-indic-dist-200M", trust_remote_code=True)
    translated_sentences = []
    for i in tqdm(range(0, len(sents), 10), desc="Batch"):
        batch = []
        for sent in sents[i:i+10]:
            tokens = indic_tokenize.trivial_tokenize(sent, lang='en')
            eng_transliterated = ItransTransliterator.to_itrans(" ".join(tokens), "en")
            batch.append(eng_transliterated)
        batch = ip.preprocess_batch(batch, src_lang=src, tgt_lang=tgt)
        batch = tokenizer(batch, src=True, return_tensors="pt")
        with torch.inference_mode():
            outputs = model.generate(**batch, num_beams=5, num_return_sequences=1, max_length=100)
        outputs = tokenizer.batch_decode(outputs, src=False)
        outputs = ip.postprocess_batch(outputs, lang=tgt)
        translated_sentences.extend(outputs)
    return translated_sentences

# Translate the English sentences to Marathi
english_to_marathi_sentences = indictranslate("eng_Latn", "mar_Deva", selected_sentences_en)
```

```python
def indictranslate(src, tgt, sents):

    tokenizer = IndicTransTokenizer(direction="indic-indic")
    ip = IndicProcessor(inference=True)
    model = AutoModelForSeq2SeqLM.from_pretrained("ai4bharat/indictrans2-indic-indic-dist-320M", trust_remote_code=True)

    translated_sentences = []

    for i in tqdm(range(0, len(sents), 10), desc="Batch"):
        batch = []
        for sent in sents[i:i+10]:
            tokens = indic_tokenize.trivial_tokenize(sent, lang='mr')
            batch.append(" ".join(tokens))

        batch = ip.preprocess_batch(batch, src_lang=src, tgt_lang=tgt)
        batch = tokenizer(batch, src=True, return_tensors="pt")

        with torch.inference_mode():
            outputs = model.generate(**batch, num_beams=5, num_return_sequences=1, max_length=100)

        outputs = tokenizer.batch_decode(outputs, src=False)
        outputs = ip.postprocess_batch(outputs, lang=tgt)

        translated_sentences.extend(outputs)

    return translated_sentences

# Translate the Marathi sentences to Hindi
marathi_to_hindi_sentences = indictranslate("mar_Deva", "hin_Deva", selected_sentences_mr)

```

##### Functions to calculate  and display scores

```python
!pip install Rouge
from nltk.translate.bleu_score import sentence_bleu
from rouge import Rouge
import os
import json
```

```python

def calculate_bleu_score(ground_truth_sentences, translated_sentences):
  bleu_scores = [sentence_bleu([ground_truth], translated) for ground_truth, translated in zip(ground_truth_sentences, translated_sentences)]
  avg_bleu_score = sum(bleu_scores) / len(bleu_scores)
  return avg_bleu_score

rouge = Rouge()
def calculate_rough_score(ground_truth_sentences, translated_sentences):
  rough_score={}
  rouge_scores = rouge.get_scores(translated_sentences, ground_truth_sentences, avg=True)
  #Extracting ROUGE-1, ROUGE-2, and ROUGE-L scores
  rouge_1_score = rouge_scores['rouge-1']['f']
  rouge_2_score = rouge_scores['rouge-2']['f']
  rouge_l_score = rouge_scores['rouge-l']['f']
  rough_score["rouge_1_score"]=rouge_1_score
  rough_score["rouge_2_score"]=rouge_2_score
  rough_score["rouge_L_score"]=rouge_l_score
  return rough_score

def display_rough_scores(rough_sc):
  for key, value in rough_sc.items():
    print(f"{key}: {value}")
```
